При этом функция $u(x,t)=g(t)f(\xi ),$ где $\xi=\frac{x}{\phi(t)}$  
1) Найти функции $g(t)\ и\ \phi(t)$  
2) Какому дифференциальному уравнению при эторм удовлетворяет функция $f(\xi)$? Решить численно получившееся уравнение для $f(\xi)$ с условиями $f'(\xi=0)_{\xi}=0;\ f(\xi=l)=0,\ l<\infty$, с дополнительными требованиями $f^{\alpha} f'_{\xi}|_{\xi=l}=0$  
3)  CравCте поведение численного решения уравнения (##) с автомодельным. (Так как известны $g(t),\ \phi(t),\ f(\xi)$, то тем самым найдено $u(x,t))$

### Пункт №1  
Уравнение (##) содержит источник тепла , степенным образом зависящий от температуры. Согласно исследованиям уравнения (##), проведенных в школе А.А. Самарского и С.П. Курдюмова, было установлено наличие так называемой “фундаментальной длины”, на которой происходит горение среды, при этом тепло не распространяется за пределы данной длины, а температура может расти неограниченно в течение конечного интервала времени. Уравнение (##) допускает специальный класс автомодельных решений, который получил название режимов с обострением

При любых  $\alpha>0$ и $\beta>1$ уравнение (##) имеет неограниченные автомодельные решения следующего вида:  
$u(t,x) = (t_f - t)^{\frac{-1}{\beta - 1}} f(\xi),\ где\ \xi = \frac{x}{(t_f - t)^q},\ где\ q = \frac{\beta - \alpha - 1}{2(\beta-1)}$  
Тогда $g(t) = (t_f - t)^{\frac{-1}{\beta - 1}}$, а $\phi(t) = \frac{1}{(t_f - t)^q}$
### Пункт №2  
Уравнение на $f(\xi)$ имеет вид: 
> $\def\dd#1#2{\frac{\partial#1}{\partial#2}} \dd{}\xi(f^{\alpha}\dd{f}\xi)- \frac1{\alpha}f(\xi) + f^{\beta}(\xi) = 0,\ где\ p = \frac{-1}{-1 + \beta}$, $q = \frac{\beta - \alpha - 1}{2(-1 + \beta)}$  

Теперь найдем аналитическое и численное решение этого уравнения взяв начальные условия описанные в описании задачи:  
$f'(\xi=0)_{\xi}=0;\\
f(\xi=l)=0,\ l<\infty\ \ \ \ \ \ \ \ \ (**)\\
f^{\alpha} f'_{\xi}|_{\xi=l}=0$

#### Аналитическое решение
$f(\xi) = [\frac{2\beta}{\alpha(\beta+1)}*cos^2(\frac{\pi x}{L})]^{\frac{1}{\alpha}},\ где\ L = \frac{2\pi( \beta + 1)^{1/2}}{\alpha}$  

##### Решение дифференциального уравнения:
Вернемся к решению дифференциального уравнению: $ u(x,t) = g(t)f(\frac{x}{\phi(t)}) = \frac1{t^p}f(x/(1/t^q))\ \Longrightarrow $ 

> $u(x,t) = (t_f - t)^{p}[\frac{2\beta}{\alpha(\beta+1)}*cos^2(\frac{\pi \xi}{L})]^{\frac{1}{\alpha}},\ где\ p = \frac{1}{1-\beta},\ q = \frac{\beta - \alpha - 1}{2(1-\beta)}$

#### Аппроксимация уравнения
Разобьем промежуток $[0,L]$ на n равных частей. Пусть $h = (L − 0)/n$ и построим сетку узлов с шагом $h:\ x_i = a + i*h,\ i = 0, 1,...,n.$ Назовем эту сетку узлов основной. Решение исходной задачи будем отыскивать в виде таблицы значений в точках сетки $f_i \approx y(x_i)$.  
Раскроем скобки у дифференциального уравнения на $f(\xi)$:  
$\frac{\partial^2 f}{\partial \xi^2} = \frac{y_{i+1} - 2y_i + y_{i-1}}{h^2}$  
  
$\frac{\partial f}{\partial \xi} = \frac{y_{i+1}-y_{i-1}}{2h}$  
  
>$-(y^{\alpha}_{i+1/2}\frac{y_{i+1}-y_i}{h^2} - y^{\alpha}_{i-1/2}\frac{y_i - y_{i-1}}{h^2}) + y_i^{\beta} - \frac1{\alpha}y_i = 0,\ i = 0, 1,...,n;$

$A_iy_{i-1} - B_iy_i + C_iy_{i+1} = 0,\ где\ i = 1,2,3,...,n-1$  
  
При это пределим $ y^{\alpha}_{i+1/2} = (\frac{y_{i+1}+y_i}{2})^{\alpha} \\ 
\ y^{\alpha}_{i-1/2} = (\frac{y_{i}+y_{i-1}}{2})^{\alpha}$

##### Исследуем на аппроксисацию
$-(y^{\alpha}_{i+1/2}\frac{y_{i+1}-y_i}{h^2} - y^{\alpha}_{i-1/2}\frac{y_i - y_{i-1}}{h^2}) + y_i^{\beta} - \frac1{\alpha}y_i = 0,\ i = 0, 1,...,n$  
$y_i = y$  
$y_{i \pm 1/2} = y_i \pm y' * h/2 $  
$y_{\pm i} = y_i \pm y'h$


$-((y^{\alpha} + C_{\alpha}^1y^{\alpha-1}y'(h/2) + C_{\alpha}^2y^{\alpha-2}y''(h/2)^2)\frac{yh}{h^2} - (y^{\alpha} + C_{\alpha}^1y^{\alpha-1}y'(h/2) + C_{\alpha}^2y^{\alpha-2}y''(h/2)^2)\frac{yh}{h^2}) + y^{\beta} -\frac1{\alpha}y = C_{\alpha-1}^1 y^{\alpha-1}y' + y^{\alpha}y'' + y^{\beta} + \frac{1}{\alpha}y + O(h^2)$  
Значит аппроксимация 2-ого порядка

#### Аппроксимация краевых условий:
$\frac{df}{dx} = \frac{y_{i+1} - y_{i-1}}{2h} \Longrightarrow y_2 = y_1$  
$y_n = 0$  
Тогда начиная с i=1:  
 1) $B_1y_1 + C_1y_2 = 0$  
 2) $A_2y_1 + B_2y_2 + C_2y_3 = 0$  
 ...  
 n-1)$ A_{n-1}y_{n-2} + B_{n-1}y_{n-1} = 0$

Заметим, что матрица системы является трехдиагональной. А значит мы можем воспользоваться методом прогонки, который как раз применяется для решения систем с трехдиагональной матрицей.  
Будем отыскивать решение системы в виде:  
>$y_i = s_iy_{n+1} + k_i$ 

Из первого уравнения системы: $s_1 = 1; k_1 = 0$.  
Для i: $s_i = \frac{C_i}{B_i −A_is_{i−1}};\ k_i = \frac{A_it_{i−1}}{B_i −A_is_{i−1}}$

#### Исследование на устойчивость

Для устойчивости метода прогонки должно выполняться следующее условие:  

>$ |c_i| >|a_i| + |b_i|\\
  \lambda_1*\lambda_2 \leqslant 1 $
  
  $ где\ \lambda_1 = 1\ и\ \lambda_2 = 0 $ значения взяты из начальных условий: $y_1 = y_0\ и\ y_N = 0$

Подифицируем функцию chis_f(x, alpha) так чтобы она выдавала нам коэффициенты $a_i,\ b_i,\ c_i$

```python
def chis_f(x, alpha):
    beta = alpha + 1
    global a
    global L
    global hx
    def anal_f(x, alpha):
        beta = alpha + 1
        global a
        global l
        global hx
        p = 1/(1.0-beta)
        q = (beta - alpha - 1)/2.0/np.abs((1.0-beta))
        return (2.0*beta/(alpha*(beta + 1))*(np.cos(np.pi*x/L))**2)**(1/alpha)
    p = 1/(1.0-np.abs(beta))
    q = (np.abs(beta) - alpha - 1.0)/2.0/(1.0-np.abs(beta))
    a = []
    b = []
    c = []
    for i in range(1, 183) :
        a.append(np.abs(-(anal_f(hx*(i), alpha) + anal_f(hx*(i-1), alpha))**(alpha)/2.0/((2*hx)**2)))
        b.append(np.abs(-(anal_f(hx*(i-1), alpha) + anal_f(hx*i, alpha))**(alpha)/2.0/((2*hx)**2) - (anal_f(hx*i, alpha))**(alpha-1) - 1/alpha - (anal_f(hx*i, alpha) - anal_f(hx*(i-1), alpha))**(alpha)/2.0/((2*hx)**2)))
        c.append(np.abs((anal_f(hx*0.1*(i+1), alpha) + anal_f(hx*(i), alpha))**(alpha)/2.0/((2*hx*0.1)**2)))
    return {'a': a,'b': b,'c': c}
```

```python
kof = chis_f(x, alpha+10.0)
```

##### Исследуем на устойчивость
Воспользуемся спектральным признаком:  
$y_i = q^i$  
Тогда $ -q^{i(\alpha+1)}((\frac{q+1}{2})^{\alpha}\frac{q + 1}{h^2} - (\frac{1 - 1/q}{2})^{\alpha} \frac{1 - 1/q}{h^2} + q^{i\beta} - \frac1{\alpha}q^i = 0$  
 Путем сложных преобразований  получим:
 >$q^{N\alpha} = \frac{h^2}{C\alpha}$, где С - константа не зависящая от qб при этом С>1. N - число узлов в решетке.

В итоге $q = (\frac{h^2}{C\alpha})^{\frac1{N\alpha} }$  
Подставляя значения, получим что q = 0.04
